# PYTHOMON

PYTHOMON is a text-based creature collector minigame inspired by Pokémon. In this game, the Pythomon Professor gives you a task to collect 5 specific Pythomon species for her research. Alongside your partner, work to complete the Pythodex as the first steps to becoming a Pythomon trainer.

GitHub Link: https://github.com/laculanding/PYTHOMON

***
If you are unfamiliar with Pokémon games, here are some mechanics to keep in mind that will not be explicitly explained in the game:
- To catch Pythomon, go to the `Route` area from the menu. Here, the player will be able to walk through a patch of "tall grass".
    - Each time the player moves, there is a set chance to encounter a wild Pythomon.
    - If this is triggered, the player enters a battle with the wild Pythomon. The opponent is randomly chosen based on the weighted probabilities of each of the available Pythomon. (Note that the rarest Pythomon's encounter rate is 10%).
    - The player will be given three options: `Attack`, `Catch`, and `Run`. The basic attack and catch mechanics will be explained in the game, but here some other things to note:
        - The battles are not turn-based. Only you will be able to make moves and the opponent will not be able to attack.
        - Unlike in the actual Pokémon games, you will only be able to use one Pythomon in battle. This is the Pythomon that the Professor gives to you. Thus, you will not be able to add the Pythomon you catch to your team. There is also no leveling system.
        - Your Pythomon partner only has one move to attack. This move randomly does between 2 to 4 damage to the opponent. You will not be able to miss nor perform critical hits.
        - If the opponent gets to 0 HP, it will faint and the battle will end.
        - You have unlimited Pythoballs. Happy catching!
        - Choosing to run will immediately end the battle. It is guaranteed every time. <br><br>
- To view your progress, go to the `Pythodex` from the menu.
    - You will be able to view information about the Pythomon you have caught here.
    - Unlike in the actual Pokémon games, simply encountering the Pythomon will not register its name in the Pythodex. You will have to catch it!
    - Be sure to use the `check Pythodex` option, especially when you've caught all the Pythomon! This is how you will be able to officially win the game.
